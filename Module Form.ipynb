```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return np.sin(x)
```

```python
def funcSecondDerivative(x_point, h):
    return (func(x_point + h) - 2 * func(x_point) + func(x_point - h)) / (h ** 2)
```

```python
n = 50
x = np.linspace(0, 2 * np.pi, n)
h = 0.1
y_derivative = np.zeros(n)
for i in range(n):
    y_derivative[i] = funcSecondDerivative(x[i], h)
```

```python
plt.plot(x, y_derivative)
```
